```python
import pandas as pd
import dill
from collections import defaultdict
from datetime import datetime, timedelta
```

```python
tickers = set(dill.load(open('dividend_tickers.pkd','rb')))
print(len(tickers))
tickers
```

```python
def get_ticker_dates(tickers):
    d_daily_data = dill.load(open('data/d_all_EOD.pkd','rb'))
    d_ticker_dates = {}

    for ticker in tickers:
        df = d_daily_data[ticker]
        dates = df[df.divCash > 0].date.values[-20:]
        dates = [datetime.strptime(str(x)[:10],'%Y-%m-%d') for x in dates]
        d_ticker_dates[ticker] = dates
    return d_ticker_dates
get_ticker_dates(tickers)
```
